```python
def level_order_traversal(root):
    """
    Args:
     root(BinaryTreeNode_int32)
    Returns:
     list_list_int32
    """
    
    result=[]
    if root==None:
        return
    helper(root,result)
    return result
def helper(root,result):
    q=deque()
    q.append(root)
    
    while q:
        level=[]
        node_count=len(q)
        while node_count>0:
            node=q.popleft()
            if node.left:
                q.append(node.left)
            if node.right:
                q.append(node.right)
            level.append(node.value)
            node_count-=1
        result.append(level)
    return result
```

```python
def height_of_binary_tree(root):
    if root is None:
        return 0

    left_subtree_height = height_of_binary_tree(root.left)
    right_subtree_height = height_of_binary_tree(root.right)

    return max(left_subtree_height, right_subtree_height) + 1

# so here the dfs(leaf nodes) gives 0 so left will be zero for leaf nodes and that is give up to the parent

# if we define the height with nodes we initialize with = 0, if it is with edges we do it with -1
def find_height(root):

    if root is None:
        return -1
    max_child_height=-1
    for child in root.children:
        max_child_height = max(max_child_height, find_height(child))
    
    return max_child_height + 1
```

```python
def find_single_value_trees(root):

    if root == None:
        return 0
    global count
    count =0
    dfs(root)
    return count
def dfs(node):
    if node.left ==None and node.right == None:
        global count
        count+=1
        return True
    unival=True
    if node.left:
        bl=dfs(node.left)
        if bl is False or (node.value!=node.left.value):
            unival= False
    if node.right:
        br=dfs(node.right)
        if br is False or (node.value!=node.right.value):
            unival= False
    if unival:
        count+=1
    return unival
    
#but using global variuable may be problematic in multi nested conditions so
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

class Solution:
    def countUnivalSubtrees(self, root: TreeNode) -> int:
        self.count = 0
        
        def is_unival(node):
            if not node:
                return True
            
            # Check left and right subtree
            left_is_unival = is_unival(node.left)
            right_is_unival = is_unival(node.right)
            
            # If left or right is not unival, return False
            if not left_is_unival or not right_is_unival:
                return False
            
            # If left child exists and its value is different, return False
            if node.left and node.left.val != node.val:
                return False
            
            # If right child exists and its value is different, return False
            if node.right and node.right.val != node.val:
                return False
            
            # If the current subtree is unival, increment the count
            self.count += 1
            return True
        
        is_unival(root)
        return self.count

#alternate
def recursive_helper(node, count):
    if not node:
        return True

    is_left_unival = recursive_helper(node.left, count)
    is_right_unival = recursive_helper(node.right, count)

    if (
        (not node.left or (is_left_unival and node.left.value == node.value))
        and (not node.right or (is_right_unival and node.right.value == node.value))
    ):
        count[0] += 1
        return True
    else:
        return False

def find_single_value_trees(root):
    count = [0]
    recursive_helper(root, count)
    return count[0]
```

```python
#Symmetric Tree
#Given a binary tree, check whether it is a mirror of itself i.e. symmetric around its centre.

def check_if_symmetric(root):

    if not root:
        return True
    def ismirror(node1,node2):
        if not node1 and not node2:
            return True
        if not node1 or not node2:
            return False
        return (node1.value==node2.value and ismirror(node1.left,node2.right) and ismirror(node1.right,node2.left))
    return ismirror(root.left, root.right)
```

```python
#utilizing inorder traversal as it traverses the tree in increasing order this has time and space complexity as O(n)
def kth_smallest_element(root, k):

    result=[]
    dfs(root,result)
    return result[k-1]
    
def dfs(node,result):
    if node is None:
        return
    
    dfs(node.left,result)
    result.append(node.value)
    dfs(node.right,result)
    
# reducing the complexity

def kth_smallest_element(root, k):

    # Variable to store the count of nodes visited during in-order traversal
    count = 0
    result = None
    
    # Helper function to perform in-order traversal
    def dfs(node):
        nonlocal count, result
        if node is None:
            return
        
        # Traverse the left subtree
        dfs(node.left)
        
        # Increment count as we visit a node
        count += 1
        
        # If count matches k, store the result
        if count == k:
            result = node.value
            return  # Stop recursion as we have found the k-th smallest
        
        # Traverse the right subtree
        dfs(node.right)
    
    # Start DFS traversal
    dfs(root)
    
    return result
```
